# DATA VISUALIZATION USING MATPLOTLIB AND SEABORN
Data visualization is the graphic representation of data. It involves producing images that communicate relationships among the represented data to viewers. Visualizing data is an essential part of data analysis and machine learning.

```python
# start by importing the libararies
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
# to ensure that graphs are shown within the jupyter notebook itself and not as pop-up windows
```

# 1. LINE CHARTS
One of the simplest and most widely used data visualization techniques. A line chart displays information as a series of markers or data points, connected by straight lines. You can customize the shape, color, size, and aesthetic elements of the markers and lines for better visual clarity.


```python
numbers = [0.5, 1.2, 0.91, 0.7, 1.9, 3.0, 2.5]
# we can visualize how these numbers change over time using a line chart. We can visualize this using the plt.plot function
plt.plot(numbers)
```

# Customizing the X axis.
We can make the data look more meaningul and informative. Let us add the year for which the data is being plotted. We can do this by 2 arguements.

```python
years = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
plt.plot(years, numbers) # 1st arguement is X axis amd 2nd arguement is Y axis
```

# Axis labels
We can add labels to each axis to show what each axis represents. It is done by usimg the plt.xlabel and plt.ylabel methods.

```python
plt.plot(years, numbers)
plt.xlabel("Years")
plt.ylabel("Numbers")
```

# Plotting multiple Lines
You can also plot multiple lines on the same graph. Just call the plt.plot function multiple times.

```python
rice = [10, 24, 50, 20, 60, 85, 100, 195, 300, 200, 75]
cement = [250, 300, 100, 150, 200, 118, 187, 99, 100, 95, 70]
year = range(2014, 2025)

plt.plot(year, rice)
plt.plot(year, cement)
plt.xlabel("Year")
plt.ylabel("Bags Supplied")
```

# Chart Title and Legend
To differentiate between multiple lines, we can include a legend within the graph using the plt.legend function and a chart title by using the plt.title function.

```python
rice = [10, 24, 50, 20, 60, 85, 100, 195, 300, 200, 75]
cement = [250, 300, 100, 150, 200, 118, 187, 99, 100, 95, 70]
year = range(2014, 2025)

plt.plot(year, rice)
plt.plot(year, cement)
plt.xlabel("Year")
plt.ylabel("Bags Supplied")
plt.title("Commodity Supplies")
plt.legend(["Rice", "Cement"])
```

# Line Markers
We can also show markers for the data points on each line using the "marker" argument of the plt.plot function. Matplotlib supports different types of markers like circle, cross, square, diamond, e.t.c

```python
rice = [10, 24, 50, 20, 60, 85, 100, 195, 300, 200, 75]
cement = [250, 300, 100, 150, 200, 118, 187, 99, 100, 95, 70]
year = range(2014, 2025)

plt.plot(year, rice, marker="o")
plt.plot(year, cement, marker="x")
plt.xlabel("Year")
plt.ylabel("Bags Supplied")
plt.title("Commodity Supplies")
plt.legend(["Rice", "Cement"])
```

# Styling Lines and Markers
The plt.plot function supports many arguments foe styling lines and markers.
* colors or c: set the color of the line
* linestyle or ls: choose between a solid or dashed line
* linewidth or lw: set the width of a line
* markersize or ms: set the size of markers
* markeredgecolor or mec: set the edge color for markers
* markeredgewidth or mew: set the edge width for markers
* markerfacecolor or mfc: set the fill colors for markers
* alpha: opacity of the plot

```python
rice = [10, 24, 50, 20, 60, 85, 100, 195, 300, 200, 75]
cement = [250, 300, 100, 150, 200, 118, 187, 99, 100, 95, 70]
year = range(2014, 2025)

plt.plot(year, rice, marker="o", c="b", ls="-", lw=2, ms=8, mew=2, mec="navy")
plt.plot(year, cement, marker="x", c="r", ls="--", lw=3, ms=10, alpha=.5)
plt.xlabel("Year")
plt.ylabel("Bags Supplied")
plt.title("Commodity Supplies")
plt.legend(["Rice", "Cement"])
```

# The "fmt" argument
It provides a shorthand for specifying the line style, marker, and line color. It can be provided as the 3rd argument to the plt.plot function.
fmt = "[marker][line][color]"

```python
rice = [10, 24, 50, 20, 60, 85, 100, 195, 300, 200, 75]
cement = [250, 300, 100, 150, 200, 118, 187, 99, 100, 95, 70]
year = range(2014, 2025)

plt.plot(year, rice, "s-b")
plt.plot(year, cement, "o--r")
plt.xlabel("Year")
plt.ylabel("Bags Supplied")
plt.title("Commodity Supplies")
plt.legend(["Rice", "Cement"])
```

# Changing the figure size
You can use the plt.figure function to change the size of the figure

```python
rice = [10, 24, 50, 20, 60, 85, 100, 195, 300, 200, 75]
cement = [250, 300, 100, 150, 200, 118, 187, 99, 100, 95, 70]
year = range(2014, 2025)

plt.figure(figsize=(12, 6)) # the size has been modified

plt.plot(year, rice, marker="o")
plt.plot(year, cement, marker="x")
plt.xlabel("Year")
plt.ylabel("Bags Supplied")
plt.title("Commodity Supplies")
plt.legend(["Rice", "Cement"])
```

# Histogram
It represents a distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall into each bin.
We can use the plt.hist function to create a histogram.

```python
plt.title("Students GPA")
plt.hist(students_df["GPA"]);
```

# We can control the size and number of bins
By using the bins argument

```python
plt.title("Students GPA")
plt.hist(students_df["GPA"], bins=5); # specifying the number of bins
```

```python
import numpy as np
plt.title("Students GPA")
plt.hist(students_df["GPA"], bins=np.arange(2, 4, 0.25)); # specifying the boundaries of each bin
```

# Multiple Histograms
Similar to line charts, we can also draw multiple histograms in a single chart. We can reduce the opacity of each histogram, so that the bars of one histogram don't hinder the others.

Let us draw seperate histogram for different Majors

```python
science_df = students_df[students_df["Major"] == "Science"]
arts_df = students_df[students_df["Major"] == "Arts"]
business_df = students_df[students_df["Major"] == "Business"]

plt.title("Students Performance Distribution")
plt.hist(science_df["GPA"], alpha = 0.5)
plt.hist(arts_df["GPA"], alpha = 0.5);
```

We can as well stack multiple histograms on top of one another

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
plt.title("Students Performance Distribution")
plt.hist([science_df["GPA"], arts_df["GPA"], business_df["GPA"]], stacked=True)
plt.xlabel("GPA")
plt.ylabel("Count of Students")
plt.legend(["Science", "Arts", "Business"])
```

# Bar Chart
They are quite similar to line charts i.e they show a sequence of values, however a bar is shown for each value rather than points connected by lines. We can use the plt.bar function to plot a bar.

```python
plt.figure(figsize=(12, 6))

sns.set_style("whitegrid") # we have added a seaborn style called whitegrid

rice = [10, 24, 50, 20, 60, 85, 100, 195, 300, 200, 75]
cement = [250, 300, 100, 150, 200, 118, 187, 99, 100, 95, 70]
year = range(2014, 2025)

plt.bar(year, rice, alpha = 0.9)
plt.bar(year, cement, alpha = 0.5) # like histogram, bar charts can also be stacked
plt.xlabel("Year")
plt.ylabel("Bags Supplied")
plt.title("Commodity Supplies")
plt.legend(["Rice", "Cement"]);
```

We can also use a line and bar chart together

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid") # we have added a seaborn style called whitegrid

rice = [10, 24, 50, 20, 60, 85, 100, 195, 300, 200, 75]
cement = [250, 300, 100, 150, 200, 118, 187, 99, 100, 95, 70]
year = range(2014, 2025)

plt.plot(year, rice, "x--r") # line chart
plt.bar(year, cement) # bar chart
plt.xlabel("Year")
plt.ylabel("Bags Supplied")
plt.title("Commodity Supplies")
plt.legend(["Rice", "Cement"]);
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid") # we have added a seaborn style called whitegrid

rice = [10, 24, 50, 20, 60, 85, 100, 195, 300, 200, 75]
cement = [250, 300, 100, 150, 200, 118, 187, 99, 100, 95, 70]
year = range(2014, 2025)

plt.bar(year, rice) 
plt.bar(year, cement, bottom=rice) # stacked bar chart
plt.xlabel("Year")
plt.ylabel("Bags Supplied")
plt.title("Commodity Supplies")
plt.legend(["Rice", "Cement"]);
```

# Heat Map
It is used to visualize a 2 dimensional data like a matrix or a table using colors.

```python
sample_df = sns.load_dataset("flights")

plt.plot(sample_df["passengers"]); # using line chart
```

A better to way do this is to use a heat map.

```python
flight_df = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")
flight_df # looks more beautiful. This is a matrix
```

```python
plt.figure(figsize=(12, 6))
plt.title("Flight Data")
sns.heatmap(flight_df);
```

```python
plt.figure(figsize=(12, 6))
plt.title("Flight Data")
sns.heatmap(flight_df, fmt="d", annot=True, cmap="Purples");
```
