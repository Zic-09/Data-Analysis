# 1. How does water intake correlate with workout performance (Session Duration and Calorie Burn)?

```python
correlation = gym_df["Water_Intake (liters)"].corr(gym_df["Session_Duration (hours)"])
print(f"Correlation Coefficient (r): {correlation}")

sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))
plt.title("Work Out Performance")
sns.scatterplot(data=gym_df, 
    x="Water_Intake (liters)", y="Session_Duration (hours)", 
    hue="Experience_Level", style="Gender", palette="colorblind", s=90);
```

# From the Scatter Plot, I Observed the following:
* As the Water Intake (liters) increases, the Session Duration (hours) does not show a significant increase. The trend indicates a very weak positive correlation (r=0.28), meaning members who drink more water do not necessarily tend to exercise longer.

# For Session Duration (hours)
* As the Water Intake (liters) increases, the Session Duration (hours) does not show a significant increase. The trend indicates a very weak positive correlation (r=0.28), meaning members who drink more water do not necessarily tend to exercise longer.

* Possible Reasons:
1. Individual Differences in Hydration Needs
* Body Composition: Individuals with different body sizes, muscle mass, or metabolic rates may have varying water needs.
* Sweat Rate: Members who sweat more may need to drink more water but might not necessarily exercise longer.
* Pre-Exercise Hydration: Members who hydrate well before exercise may require less water during the session, regardless of its duration.
2. Other Dominant Factors Affect Session Duration
* Experience Level: Experienced members likely exercise longer due to better stamina and training, regardless of water intake.
* Workout Intensity: High-intensity workouts may be shorter but require more water to compensate for sweat loss.
* Fitness Goals: Members focused on quick, targeted workouts may drink less water but still achieve their session objectives.
3. Timing of Water Consumption
* Pre- vs. During-Exercise Intake: The dataset did not differentiate between water consumed before the session versus during it, which affects correlation.
* Post-Workout Hydration: Members might hydrate extensively after their session, not during it, diluting the relationship between intake and duration.
4. Behavioral or Environmental Factors
* Availability of Water: Members may drink more due to convenient access to hydration stations, not necessarily because they exercise longer.
* Social Habits: Some members may take water breaks more frequently as a habit or to socialize, rather than due to the demands of longer sessions.
5. Measurement Limitations
* Self-Reported Data: If water intake is self-reported, inaccuracies could weaken the correlation.
* Granularity of Data: If water intake is not recorded in real-time or session duration includes rest periods, the true relationship may be masked.

# For Calories Burned
* Water Intake (liters) does not lead to an increase in Calories Burned. This trend indicates a Weak Positive Correlation (r=0.36). This means that members who drank more water did not necessarily burn more calories. 

* Possible Reasons:
1. Calorie Burn Depends on Physical Activity
* Exercise Intensity and Type: Calories burned are primarily determined by the intensity and type of exercise. For example:
    High-intensity workouts burn more calories in less time.
    Strength training may burn fewer calories immediately but have a lasting metabolic effect.
2. Duration of Exercise: A longer session might burn more calories, but water intake may not directly drive this.


# 2. Are members with high BMI improving over time in terms of fat percentage and calorie burn?

* I cannot get an accurate answer from the dataset with this question. I need the data that contains a before session and after session including the timestamps. This will help me track accuartely their improvement over time.

```python
correlation = gym_df["BMI"].corr(gym_df["Fat_Percentage"])
print(f"Correlation Coefficient (r): {correlation}")

sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
plt.title("Members BMI Improvement by Fat Percentage")
sns.scatterplot(data=gym_df, x="BMI", y="Fat_Percentage", hue="Gender", palette="colorblind");
```

```python
# Want to get percentage of women who had above 30% fat
threshold = 30
threshold_2 = 25
threshold_3_min = 18
threshold_3_max = 24 
females = gym_df[gym_df["Gender"] == "Female"]
females_percentage = females[females["Fat_Percentage"] > threshold]
percentage = (len(females_percentage) / len(females)) * 100
print(f"Percentage of ladies with fat above {threshold}: {percentage:.2f}%")

# for those with above 30% fat and above/equal 25 BMI
females_2 = gym_df[gym_df["Gender"] == "Female"]
females_percentage_2 = females_2[(females_2["Fat_Percentage"] > threshold) & (females_2["BMI"] >= threshold_2)]
percentage_2 = (len(females_percentage_2) / len(females_2)) * 100
print(f"Percentage of ladies with fat above {threshold} and BMI equal & above {threshold_2}: {percentage_2:.2f}%")

# males who have high bmi (25 & above) yet low fat percentage (range of 18-24%)
males = gym_df[gym_df["Gender"] == "Male"]
male_percent = males[(males["BMI"] >= threshold_2) & (males["Fat_Percentage"] >= threshold_3_min)
                    & (males["Fat_Percentage"] <= threshold_3_max)]
perc_male = (len(male_percent) / len(males)) * 100
print(f"Percentage of males with BMI equal & above {threshold_2} amd fat% within {threshold_3_min}-{threshold_3_max}: {perc_male:.2f}%")
```

```python
correlation = gym_df["BMI"].corr(gym_df["Calories_Burned"])
print(f"Correlation Coefficient (r): {correlation}")

sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
plt.title("Members Improvement")
sns.scatterplot(data=gym_df, x="BMI", y="Calories_Burned");
```
