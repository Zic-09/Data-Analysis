# Movie (ID) with the highest and lowest views
* ID 228 had the highest view (670990)
* ID 53 had the lowest views (2461)

```python
plt.figure(figsize=(12, 6))
plt.title("Movie with the highest views")
sns.lineplot(data=cleaned_ytube, x="ID", y="Views", marker="o")
```

# Year with highest and lowest movies
* 2017 had the highest movie (126)
* 2023 had the lowest movies(10)

```python
count_years = cleaned_ytube.groupby("Year")["ID"].count().reset_index() # set to count the number of movies released in each year

sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
plt.title("Year with the highest movies")
sns.barplot(data=count_years, x="Year", y="ID", hue="Year", legend=False, palette="viridis")
```

# Which month had the highest and Lowest Likes in Year 2017?
* In July, there were 119,013 Likes recorded, making it the month in year 2017 with the highest Likes.
* In January, there were 4,823 Likes recorded, making it the month in year 2017 with the lowest Likes.

```python
subscribers = cleaned_ytube[cleaned_ytube["Year"] == 2017]
cnt_subscribers = subscribers.groupby("Month")["Likes"].sum().reset_index()

plt.title("Likes in Year 2017")
sns.lineplot(data=cnt_subscribers, x="Month", y="Likes", marker="o")
```

```python
dis_subs = subscribers.groupby("Month")["Dislikes"].sum().reset_index()

plt.title("Dislikes in Year 2017")
sns.lineplot(data=dis_subs, x="Month", y="Dislikes", marker="o")
```

# Month with the highest and lowest views in Year 2020
* January had the highest views with 348,175 views

```python
filt_year = cleaned_ytube[cleaned_ytube["Year"] == 2017]
highest_mnth = filt_year.groupby("Month")["Views"].sum().reset_index()
plt.title("Views in Year 2017")
sns.lineplot(data=highest_mnth, x="Month", y="Views", marker = "o")
```

```python
yr_2020 = cleaned_ytube[cleaned_ytube["Year"] == 2020]
filt_2020 = yr_2020.groupby("Month")["Views"].sum().reset_index()
filt_2020
sns.set_style("whitegrid")
plt.title("Views in Year 2020")
sns.lineplot(data=filt_2020, x="Month", y="Views", marker="o")
```
