```python
import pymysql

try:
    # Connect to MySQL
    connection = pymysql.connect(
        host="localhost",
        user="root",
        password="Oluwadarasimi18",
        database="tutorial",
        port=3306
    )

    print("Connected to MySQL Database")

    # Create a cursor object
    with connection.cursor() as cursor:
        # Execute query
        cursor.execute("SELECT * FROM us_housing_data;")

        # Fetch all results
        results = cursor.fetchall()

        # Print each row
        for row in results:
            print(row)

except pymysql.MySQLError as e:
    print("Error:", e)

finally:
    if connection:
        connection.close()
        print("MySQL connection closed.")
```
